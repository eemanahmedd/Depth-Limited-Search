# Depth Limited Search

```python
# Defining graph

graph = {
    'S' : ['A', 'B'],
    'A' : ['C', 'D'],
    'B' : ['I', 'J'],
    'C' : ['E', 'F'],
    'D' : ['G'],
    'I' : ['H'],
    'J' : ['K'],
    'E' : [],
    'F' : [],
    'G' : ['L'],
    'H' : [],
    'K' : ['M'],
    'M' : [],
    'L' : []
}
```

```python
def dls(start,goal,path,level,maxDepth):
    print('Level-->',level)
    print('Testing Goal Node for ',start)
    path.append(start)
    if start == goal:
        print("Goal Node test successful")
        return path
    print('Goal Node testing failed')
    if level==maxDepth:
        return False
    print('\nExpanding the current node',start)
    for child in graph[start]:
        if dls(child,goal,path,level+1,maxDepth):
            return path
    return False
```

```python
path = list()

#user inputs
goal_node = input("Enter the goal node: ")
limit = input("Enter the limit for search: ")


result = dls('S', goal_node, path, 0, limit)
if(result):
    print("\nPath to Goal Node found")
    print("Path",path)
else:
    print("No Path available for the Goal Node in given depth limit")
```
